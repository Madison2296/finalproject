```python
def names_times(filename):
    names_times_dict = {}
    fobj = open(filename, "r", encoding = "UTF-8")#open the file
    all_records = fobj.readlines()[16:]#exclude any header information
    for line in all_records:
        record_elements = line.split(",")#create a list out of line elements
        name = record_elements[5]#extract species names
        time = record_elements[11]#extract species time period
        names_times_dict[name] = time#make a dictionary out of species names and time periods
    return names_times_dict
```
