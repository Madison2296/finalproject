## Exploring the timeline of fossil records

#### Manipulating the data

```
def names_times(filename):
    names_times_dict = {}
    fobj = open(filename, "r", encoding = "UTF-8")#open the file
    all_records = fobj.readlines()[16:] #exclude any header information
    for line in all_records:
        record_elements = line.split(",") #create a list out of line elements
        name = record_elements[5] #extract species names
        time = record_elements[11] #extract species time period
        names_times_dict[name] = time #make a dictionary out of species names and time periods
    return names_times_dict
    
#use this code, but change for my own data to explore how long a species was present
#also eliminate extra commas and entries with only a genus or family name, but not a species name
infile = "cleaned_canids.csv"
with open(infile, 'r') as ff:
    canid_recs = ff.readlines() #open the file
from collections import defaultdict
species_ranges = defaultdict(list) #use defaultdict to make dictionary able to hold a list as a value
#make a dictionary
for line in canid_recs:
    species = line.split(",")[5] #extract species names
    minage = line.split('","')[10] #extract min age
    maxage = line.split('","')[11] #extract max age
    mean_age = (float(minage) + float(maxage)) / 2 #find average age
    species_ranges[species].append(mean_age) #append to a dictionary
for key in species_ranges.keys():
    ages = species_ranges[key]
    minage = min(ages)
    maxage = max(ages)
    genus = key.split(" ")[0]
    species = key
    outline = "{},{},{},{}\n".format(genus,species,minage,maxage) #make a csv file of species and their age range
```

## Exploring the location of fossil records

#### manipulating data

```
def name_plate(filename):
    names_plates_dict = {}
    fobj = open(filename, "r", encoding = "UTF-8")#open the file
    all_records = fobj.readlines()[16:]#exclude any header information
    for line in all_records:
        record_elements = line.split(",")#create a list out of line elements
        name = record_elements[5]#extract species names
        plate = record_elements[-1]#extract species plate location
        names_plates_dict[name] = plate#make a dictionary out of species names and plate locations
    return names_plates_dict
    
def name_state(filename):
    names_states_dict = {}
    fobj = open(filename, "r", encoding = "UTF-8")#open the file
    all_records = fobj.readlines()[16:]#exclude any header information
    for line in all_records:
        record_elements = line.split(",")#create a list out of line elements
        name = record_elements[5]#extract species names
        state = record_elements[20]#extract species state location
        names_states_dict[name] = state#make a dictionary out of species names and plate locations
    return names_states_dict
```
